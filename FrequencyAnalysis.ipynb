```python
def frequency_count(data, n=1):
    # Erstellen eines Wörterbuchs, um die Häufigkeiten zu speichern
    freq_dict = {}

    # Zählen der Vorkommen jeder Buchstabenkombination der Größe n
    for i in range(len(data) - n + 1):
        item = data[i:i+n]
        if item in freq_dict:
            freq_dict[item] += 1
        else:
            freq_dict[item] = 1

    # Berechnung der relativen Häufigkeiten
    total_count = sum(freq_dict.values())
    relative_freq_dict = {k: v / total_count for k, v in freq_dict.items()}

    return freq_dict, relative_freq_dict
```

```python
import re

def read_and_clean_text(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    
    # Alle Sonderzeichen und Leerzeichen entfernen, nur Großbuchstaben beibehalten
    cleaned_text = re.sub(r'[^A-Z]', '', text.upper())
    
    return cleaned_text

cleaned_string = read_and_clean_text('CIPHER.txt')
```

```python
tolerance = 0.005

def combine_frequencies(frq_rel_origin, frq_rel_cmp):
    combined_dict = {}
    for k_origin in frq_rel_origin:
        for k_cmp in frq_rel_cmp:
            if (frq_rel_origin[k_origin] > frq_rel_cmp[k_cmp] - tolerance) and (frq_rel_origin[k_origin] < frq_rel_cmp[k_cmp] + tolerance):
                if k_origin not in combined_dict:
                    combined_dict[k_origin] = []
                combined_dict[k_origin].append(k_cmp)
    return combined_dict
```

```python
# Ersetzt auch bereits ersetzte Strings aus dem Dict. Aber wie verhindere ich das? 
from copy import deepcopy

def get_longest_value(combinde_dict):
    longest_value = 1
    for k in combinde_dict:
        if len(combinde_dict[k]) > longest_value:
            longest_value = len(combinde_dict[k])
    return longest_value

def change_cmp_string(string: str, combined_dict, n=1):
    results = [deepcopy(string)]
    for k in combined_dict:
        if len(combined_dict[k]) > len(results):
            results.append(deepcopy(results[-1]))
        for v in combined_dict[k]:
            for i in range(len(results)):
                for index in range(len(results[i])):
                    # übersetzen von k in v
                    if results[i][index:index+n] == k:
                        results[i] = results[i][:index] + v + results[i][index+n:]
    return results
```

```python
def group_in_blocks(string, block_size):
    # Teile den String in Blöcke der gewünschten Größe
    blocks = [string[i:i+block_size] for i in range(0, len(string), block_size)]
    # Füge die Blöcke mit Leerzeichen zusammen
    grouped_string = ' '.join(blocks)
    return grouped_string
```

```python
cipher = read_and_clean_text('CIPHER.txt')
encrypted_1 = change_cmp_string(cipher, combined_1, 1)

cipher_b = group_in_blocks(cipher, 5)

print(combined_1)
print("Encrypted:")

for i in range(len(encrypted_1)):
    print(group_in_blocks(encrypted_1[i], 5))

print("Original:")
print(cipher_b)
```

```python
def find_common_substrings(strings, substring_length):
    # Extrahiere alle Substrings der angegebenen Länge aus jedem String
    substrings_sets = []
    for string in strings:
        substrings = {string[i:i+substring_length] for i in range(len(string) - substring_length + 1)}
        substrings_sets.append(substrings)
    
    # Finde die gemeinsamen Substrings in allen Sets
    common_substrings = set.intersection(*substrings_sets)
    
    return common_substrings
```
